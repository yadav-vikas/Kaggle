Over time, which states produce the most honey? Which produce the least? Which have experienced the most change in honey yield?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('honeyproduction.csv')
df.head()
```

```python
states_production = df.groupby('state').totalprod.sum().reset_index()
f, ax = plt.subplots(figsize=(20, 14))
sns.barplot(x='state',y="totalprod", data=states_production, color="c")
```

```python
states_production.max()
```

```python
states_stock = df.groupby('state').stocks.sum().reset_index()
f, ax = plt.subplots(figsize=(20, 14))
sns.barplot(x='state',y="stocks", data=states_stock, color="c")
```

```python
state_year = df.groupby(['state','year','totalprod']).sum()
state_year.head()
```

```python
state_year.state.nunique().count()

```
